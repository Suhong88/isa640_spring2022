# Chapter 4: Introduction to Seaborn Library

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

plt.rcParams["figure.figsize"] = [10,8]

# load an existing dataset from seabron libraries
#tips_data = sns.load_dataset('tips')

# altenatively, you may also load your own dataset

# tips_data=pd.read_csv('data/tips.csv')

# you can also directly read from a url

tips_data=pd.read_csv('https://raw.githubusercontent.com/Suhong88/isa640_spring2022/main/Data/tips.csv')

tips_data.head()
```

### 4.2. Dist Plot

```python
plt.rcParams["figure.figsize"] = [10,8]
sns.displot(data=tips_data, x='total_bill', kde=True)
```

```python
sns.displot(data=tips_data,x="size",y="total_bill")
```

```python
sns.displot(data=tips_data,x="size",y="total_bill",kind="kde")
```

```python
sns.displot(data=tips_data,x="total_bill",kind="kde",hue="sex")
```

```python
sns.displot(data=tips_data,x="total_bill",kind="kde",hue="day")
```

```python
sns.displot(data=tips_data, x="total_bill", hue="smoker", col="sex", kind="kde")
```

### 4.3. Joint Plot

```python
sns.jointplot(x='total_bill', y='tip', data=tips_data)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips_data, hue="sex")
```

sns.jointplot(x='total_bill', y='tip', data=tips_data, kind='hex')

```python
sns.jointplot(x='size', y='total_bill', data=tips_data, kind = 'reg')
```

### 4.4. Pair Plot

```python
sns.pairplot(data=tips_data)
```

```python
sns.pairplot(data=tips_data, hue = 'time')
```

### 4.5. Rug Plot

```python
sns.rugplot(tips_data['total_bill'])
```

### 4.6. Bar Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = [8,6]
sns.set_style("darkgrid")

titanic_data = sns.load_dataset('titanic')

titanic_data.head()
```

```python
sns.barplot(x='pclass', y='age', data=titanic_data)
```

```python
sns.barplot(x='pclass', y='age', hue ='sex', data=titanic_data)
```

```python
sns.catplot(x="pclass", y="age", hue="sex", col="survived", data=titanic_data, kind="bar");
```

### 4.7. Count Plot

```python
sns.countplot(x='pclass', data=titanic_data)
```

```python
sns.countplot(x='pclass', hue='sex', data=titanic_data)
```

### 4.8. Box Plot

```python
sns.boxplot(x=titanic_data["fare"])
```

```python
sns.boxplot(y=titanic_data["fare"])
```

```python
sns.boxplot(x='alone', y='age', data=titanic_data)
```

```python
sns.boxplot(x='alone', y='age', hue='sex',data=titanic_data)
```

```python
sns.boxplot(x='alone', y='age', hue='sex',data=titanic_data, showfliers=False)
```

### 4.9. Violin Plot

```python
sns.violinplot(x='alone', y='age', data=titanic_data)
```

```python
sns.violinplot(x='alone', y='age', hue='sex',data=titanic_data)
```

```python
sns.violinplot(x='alone', y='age', hue='sex',data=titanic_data, split=True)
```

### 4.10. Strip Plot

```python
sns.stripplot(x='alone', y='age', data=titanic_data)
```

```python
sns.stripplot(x='alone', y='age', hue='sex',data=titanic_data)
```

```python
sns.stripplot(x='alone', y='age', hue='sex',data=titanic_data, dodge = True)
```

### 4.11. Swarm Plot

```python
sns.swarmplot(x='alone', y='age', data=titanic_data)
```

```python
sns.swarmplot(x='alone', y='age', hue='sex',data=titanic_data)
```

```python
sns.swarmplot(x='alone', y='age', hue='sex',data=titanic_data, dodge = True)

plt.savefig('Figures/swamplot.png')
```
